#Drop Target Variable

```python
load_features=load_data.drop(['L(T)'], axis=1)
load_features.head()
```

# Data Normaliztion

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(load_features)
X_train_df = pd.DataFrame(df_scaled, columns=load_features.columns)
X_train_df
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
n_features=10
# 12, 11, 10, 9, 8, 7, 6, 7, 8, 9, 10, 11, 12
```

```python
# encoder
encoder_input = Input(shape=(n_features), name='encoder_input')
encoder_layer1 = Dense(9, activation='relu', name='encoder_layer1')(encoder_input)
encoder_layer2 = Dense(8, activation='relu', name='encoder_layer2')(encoder_layer1)
encoder_layer3 = Dense(7, activation='relu', name='encoder_layer3')(encoder_layer2)
encoder_layer4 = Dense(6, activation='relu', name='encoder_layer4')(encoder_layer3)
encoder_layer5 = Dense(5, activation='relu', name='encoder_layer5')(encoder_layer4)
latent_space = Dense(4, activation='relu', name='latent_space')(encoder_layer5)
```

```python
#decoder
decoder_layer1 = Dense(5, activation='relu', name='decoder_layer1')(latent_space)
decoder_layer2 = Dense(6, activation='relu', name='decoder_layer2')(decoder_layer1)
decoder_layer3 = Dense(7, activation='relu', name='decoder_layer3')(decoder_layer2)
decoder_layer4 = Dense(8, activation='relu', name='decoder_layer4')(decoder_layer3)
decoder_layer5 = Dense(9, activation='relu', name='decoder_layer5')(decoder_layer4)
output = Dense(n_features, activation='sigmoid', name='Output')(decoder_layer5)
```

```python
autoencoder = Model(encoder_input, output, name='Autoencoder')

autoencoder.compile(optimizer='adam',
                    loss=tf.keras.losses.mean_squared_error)
autoencoder.summary()
```

```python
history = autoencoder.fit(X_train_df,
                          X_train_df,
                          epochs=100,
                          steps_per_epoch=10,
                          verbose=1)
```

```python
# only the encoder part
encoder = Model(inputs=encoder_input, outputs=latent_space)
#encoder.compile(optimizer='adam',loss=tf.keras.losses.mean_squared_error)
X_train_encode = encoder.predict(X_train_df)
encoded_train = pd.DataFrame(X_train_encode,
                             columns=[f"X{x}" for x in range(1, X_train_encode.shape[1]+1)])
```
