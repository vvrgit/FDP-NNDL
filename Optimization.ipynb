# **Gradient Descent Optimization Algorithm**

```python
import numpy as np
x_o=np.random.uniform(-10,10)
eta=0.9
eps=0.00001
max_iter=1000
iter=0

def derivative(x):
  x_derivative=2*x
  return x_derivative

while iter<max_iter:
  del_x=-eta*derivative(x_o)
  x_o=x_o+del_x
  iter=iter+1
  #print("iteration=",iter,"\n x value=",x_o)

print("solution x_o = ", round(x_o))


```

Min f(x)=x^2+5
f(x) is objective function
x is a control variable

Step1: Intialization of control variable (x) randomly, Initialize learning rate (eta), Initilize maximum number of iteration/epochs (iter_max)

```python
import numpy as np
x=np.random.uniform(-100,100)
eta=0.9
iter_max=200
```

```python
iter=1
while iter<=iter_max:
  slope=2*x
  step_length=-eta*slope
  x=x+step_length
  iter=iter+1
print("control variable=",round(x))
obj=x*x+5
print("objective function=",obj)
```

Multi Varibale Optimization Using Gradient Descent Algorithm

Minimize x^2+y^2+10

Step1: Initialization

```python
import numpy as np
import math as mt
eta=0.1
x=np.random.uniform(-10,10)
y=np.random.uniform(-10,10)
itermax=100
iter=1
```

Iterative Algorithm

```python
while iter<=itermax:
  delx=-eta*2*x
  dely=-eta*2*y
  x=x+delx
  y=y+dely
  iter=iter+1
```

Print Solution

```python
obj=mt.pow(x,2)+mt.pow(y,2)+10
print('Minimum value of objective function is',obj,'will get at x=',x,'y=',y)
```

# **Data Split**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data.X, data.Y, test_size = 0.20, random_state = 10)
```

```python
x_train
```

# **Initialization of ML model parameters**

```python
m=random.uniform(-10,10)
c=random.uniform(-10,10)
eta=0.9
max_iter=10000
```

# **Gradient Calculation**

```python
def gradient(m,c,x,y):
  m_deriv=-1*(y-m*x-c)*x
  c_deriv=-1*(y-m*x-c)
  return m_deriv,c_deriv
```

# **Training Using Gradient Descent Optimizer**

```python
for iter in range(max_iter):
  for sample in range(x_train.shape[0]):
    del_m,del_c=gradient(m, c, x_train.iloc[sample], y_train.iloc[sample])
    m=m-eta*del_m
    c=c-eta*del_c
print('model parameter m and c',m,c)
```

# **Model Deployment**
